```python
import numpy as np
import pandas as pd
import missingno as msno
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
%matplotlib inline
```

```python
df = pd.read_csv('/content/menu.csv')
df.head(5)
```

```python
#Check for missing value
msno.matrix(df)
```

```python
df.isna().sum()
```

```python
#Check Unique in Category Column
df.Category.unique()
```

```python
#Count for each Category in the Menu analysis visualization
plt.style.use('default')
plt.figure(figsize = (7,4), edgecolor = '0.1', dpi = 100)
a = sns.countplot(x = 'Category', facecolor = 'darkgreen', data = df)

xticks = plt.xticks(rotation = 55, family = 'serif')
yticks = plt.yticks(family = 'serif')


plt.xlabel(df['Category'].all())
plt.ylabel('Count')

a.spines['bottom'].set_color('gray')
a.spines['left'].set_color('gray')

sns.despine()
```

```python
#Number of Items per Category analysis visualization
px.density_heatmap(x='Category', y = 'Item', data_frame = df, width = 900)
```

```python
# Analysis of Nutrients
# For Calories
calories = pd.DataFrame (df.groupby('Category')['Calories'].mean())

colors = ['gray']*9
colors[3] = '#AF0038'
fig = go.Figure (data = [go.Bar(
    x = calories.index,
    y = calories['Calories'],
    marker_color = colors
)])
fig.update_layout(width = 700, height =500)
fig.update_yaxes(title = 'Avg Calories')
fig.update_xaxes(title = 'Categories')
fig.show()
```

```python
# For Fat
fat = pd.DataFrame (df.groupby('Category')['Total Fat'].mean())

colors = ['gray']*9
colors[2] = '#AF0038'
fig = go.Figure (data = [go.Bar(
    x = fat.index,
    y = fat['Total Fat'],
    marker_color = colors
)])
fig.update_layout(width = 700, height =500)
fig.update_yaxes(title = 'Avg Total Fat')
fig.update_xaxes(title = 'Categories')
fig.show()
```

```python
# For Carbohydrates
carbohydrates = pd.DataFrame (df.groupby('Category')['Carbohydrates'].mean())

colors = ['gray']*9
colors[7] = '#AF0038'
fig = go.Figure (data = [go.Bar(
    x = carbohydrates.index,
    y = carbohydrates['Carbohydrates'],
    marker_color = colors
)])
fig.update_layout(width = 700, height =500)
fig.update_yaxes(title = 'Avg Carbohydrates')
fig.update_xaxes(title = 'Categories')
fig.show()
```

```python
# Analyzing how Carbohydrates and Sugars are related

plt.style.use ('ggplot')
plt.figure(figsize = (8,5), dpi = 80)
sns.scatterplot (x = 'Carbohydrates', y = 'Sugars', data = df, alpha = 0.8, s = 60)
plt.title ('Carbohydrates vs Sugars', x =0.5, y=1.05)
```

```python
#Average Nutrient Analysis (% Daily Value) for each of the Categories

cols = ['Total Fat (% Daily Value)', 'Saturated Fat (% Daily Value)', 'Cholesterol (% Daily Value)', 'Sodium (% Daily Value)',
        'Carbohydrates (% Daily Value)', 'Dietary Fiber (% Daily Value)', 'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)',
        'Calcium (% Daily Value)', 'Iron (% Daily Value)']

table = df.pivot_table (columns ='Category', values = cols)
table.style.background_gradient(cmap = 'viridis')
```

```python
#For each Category below, we analyze the items with highest amount of Nutrients  in the Menu
#Breakfast

breakfast = df[df['Category'] == 'Breakfast']
cols = ['Calories', 'Total Fat', 'Carbohydrates', 'Sodium', 'Protein', 'Dietary Fiber']

for n in cols:
  print (n)
  a = breakfast.groupby('Item')[n].max().sort_values(ascending = False).head(1)
  print (a)
  print ('-'*50)
  print ('\n')
```
